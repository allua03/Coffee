```python
import pandas as pd

df = pd.read_csv('/content/coffee_sales_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df['День недели'],df['Количество Продаж'],color='pink')
plt.title('Total Coffee Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Coffee Sales')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x='Тип Кофе', y='Цена за Чашку',data=df, palette='Set2')
plt.title('Price Distribution for coffee')
plt.xlabel('Type of coffee')
plt.ylabel('Price')
plt.show()
```

```python
weekends = ['Saturday','Sunday']
weekdays = ['Monday','Tuesday','Wednesday','Thursday','Friday']

sales_weekday= df.loc[df['День недели'].isin(weekdays)]['Количество Продаж'].sum()
sales_weekends = df[df['День недели'].isin(weekends)]['Количество Продаж'].sum()

sales_data = {'Weekdays': sales_weekday, 'Weekends': sales_weekends}

plt.figure(figsize=(6, 4))
plt.bar(sales_data.keys(), sales_data.values(), color=['lightblue', 'pink'])
plt.title('Total Coffee Sales: Weekdays vs Weekends')
plt.xlabel('Day Type')
plt.ylabel('Total Sales')
plt.show()
```
